**K-Neirest Neighbour (KNN)**

Lazy learning machine learning algorithms - where new algorithms will learn when they have to infer classes or values from new data.

K-nearest neighbors (KNN) algorithm is a type of supervised ML algorithm which can be used for both classification as well as regression predictive problems.

```
path = "/content/drive/My Drive/classification.csv"
dataset = pd.read_csv(path)
dataset.head()
```

pre-processing data

```
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values


```

show the X variable

```
X[1:5,:]
```

show the y variable

```
y[1:5]
```

Divide the dataset into 60% training data and 40% testing

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40)
```

Show X train and X test before standardization

```
X_train[1:5,:]
```

```
X_test[1:5,:]
```

Standardization of scale

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Display standardized X_train and X_test

```
X_train[1:5,:]
```

```
X_test[1:5,:]
```

Now starts the KKN here

for example we choose k = 5

```
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X_train, y_train)
```

use the model that has been made to predict the estimated value of y - from the test data

```
y_pred = classifier.predict(X_test)
```

View results and display evaluations for good or bad classification results using a configuration matrix and accuracy


```
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```
